```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Retail Sales Big Data Analysis") \
    .getOrCreate()

spark
```

```python
df = spark.read.csv(
    "train.csv",
    header=True,
    inferSchema=True
)
```

```python
df.show(5)
```

```python
# Understand the Data
#Structure of the Data
df.printSchema()
```

```python
print("Total number of records:", df.count())
```

```python
#Data Cleaning
#Remove Null Values
df_clean = df.dropna()
```

```python
print("Records after cleaning:", df_clean.count())
```

```python
#Performing EDA
df_clean.describe().show()
```

```python
df.show()
```

```python
from pyspark.sql.functions import sum

df_clean.groupBy("Category") \
    .agg(sum("Sales").alias("Total_Sales")) \
    .orderBy("Total_Sales", ascending=False) \
    .show()
```

```python
from pyspark.sql.functions import sum

df_clean.groupBy("Product Name") \
    .agg(sum("Sales").alias("Total_Sales")) \
    .orderBy("Total_Sales", ascending=False) \
    .show(10)
```

```python
from pyspark.sql.functions import avg

df_clean.groupBy("Region") \
    .agg(avg("Sales").alias("Average_Sales")) \
    .show()
```
